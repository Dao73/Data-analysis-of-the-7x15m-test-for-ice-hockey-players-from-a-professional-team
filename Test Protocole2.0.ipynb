```python
import pandas as pd

csv_file_path = "C:\Python\Dossier csv\_Dossier RSA 7X15M 2.0 - DATA.csv.csv"

# Charger le fichier CSV dans un DataFrame
df = pd.read_csv(csv_file_path)

print(df)


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suppose que votre dataframe s'appelle df
# Supposons galement que les noms de session sont 'Session 1' et 'Session 2'

# Filtrer les donnes pour obtenir les scores de la session 1 et de la session 2
session1_data = df[df['Session'] == 1]
session2_data = df[df['Session'] == 2]

# Crer le graphique
plt.figure(figsize=(10,7 ))

# Tracer les scores pour la session 1
plt.scatter(session1_data['Sprint'], session1_data['Score'], color='blue', label='Session 1')

# Tracer les scores pour la session 2
plt.scatter(session2_data['Sprint'], session2_data['Score'], color='red', label='Session 2')

# Ajouter des tiquettes et un titre
plt.xlabel('Sprint')
plt.ylabel('Score')
plt.title('Scores par sprint pour la session 1 et la session 2')
plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculer la moyenne de chaque sprint pour chaque session
sprint_means = df.groupby(['Session', 'Sprint'])['Score'].mean().unstack()

# Crer le graphique
plt.figure(figsize=(10, 6))
for session in sprint_means.index:
    plt.plot(sprint_means.columns, sprint_means.loc[session], marker='o', label=f'Session {session}')

# Ajouter des tiquettes et un titre
plt.xlabel('Sprint')
plt.ylabel('Score moyen')
plt.title('Moyenne des scores par sprint pour chaque session')
plt.legend()
plt.grid(True)
plt.show()
```
